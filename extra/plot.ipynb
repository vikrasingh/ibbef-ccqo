```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,30)
y=x**2
plt.plot(x,y, label  = 'Distance')
z=12*x-36
plt.plot(x[10:25],z[10:25], label = 'Tangent at (6, 36)')
#w=2*np.ones(30)
#plt.plot(x,w, label ='Acceleration')
plt.xlabel('Time [hours]')
plt.ylabel('Distance [miles]')
plt.title('Distance over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,30)
z=60*x
plt.plot(x,z, label = '')
plt.vlines(6, 250,360, color = 'red', linestyles = 'dashed')
plt.hlines(250, 4.2, 6, color = 'red', linestyles = 'dashed')
plt.xlabel('Time [hours]')
plt.ylabel('Distance [miles]')
plt.title('Distance over Time')
plt.show()
```
